```python
#31. Write a program to generate a list of prime numbers using the Sieve of Eratosthenes algorithm and list comprehension.
upper_limit = int(input("Enter the upper limit for prime numbers: "))
is_prime = [True] * (upper_limit + 1)
is_prime[0], is_prime[1] = False, False  # 0 and 1 are not prime numbers

for num in range(2, int(upper_limit**0.5) + 1):
    if is_prime[num]:
        is_prime[num*num:upper_limit+1:num] = [False] * len(range(num*num, upper_limit+1, num))

prime_numbers = [num for num in range(2, upper_limit + 1) if is_prime[num]]
print(f"Prime Numbers up to {upper_limit}: {prime_numbers}")

#32. Create a program that generates a list of all Pythagorean triplets up to a specified limit using list comprehension.
limit = int(input("Enter the upper limit for Pythagorean triplets: "))
pythagorean_triplets = [(a, b, c) for a in range(1, limit+1) for b in range(a, limit+1) for c in range(b, limit+1) if a**2 + b**2 == c**2]

print(f"Pythagorean Triplets up to {limit}:", pythagorean_triplets)

#33. Develop a program that generates a list of all possible combinations of two lists using list comprehension.
list1 = [1,2,3]
list2 = ['a','b','c']

combinations = [(x, y) for x in list1 for y in list2]

print("List 1:", list1)
print("List 2:", list2)
print("Combinations:", combinations)

#34. Write a program that calculates the mean, median, and mode of a list of numbers using list comprehension.
numbers_list = [5,2,8,2,7,3,9,2,7]

mean = sum(numbers_list) / len(numbers_list)

sorted_numbers = sorted(numbers_list)
middle_index = len(numbers_list) // 2
median = (sorted_numbers[middle_index] + sorted_numbers[~middle_index]) / 2

# Calculate mode
counts = {}
for num in numbers_list:
    if num in counts:
        counts[num] += 1
    else:
        counts[num] = 1

max_count = max(counts.values())
mode = [num for num, count in counts.items() if count == max_count]

print("Numbers List:", numbers_list)
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)

#35. Create a program that generates Pascal's triangle up to a specified number of rows using list comprehension.
def generate_pascals_triangle(num_rows):
    triangle = [[1]]
    
    for i in range(1, num_rows):
        row = [1]
        row.extend([triangle[i-1][j] + triangle[i-1][j+1] for j in range(len(triangle[i-1]) - 1)])
        row.append(1)
        triangle.append(row)
    
    return triangle
num_rows = int(input("Enter the number of rows for Pascal's triangle: "))
pascals_triangle = generate_pascals_triangle(num_rows)
for row in pascals_triangle:
    print(row)
    
#36.Develop a program that calculates the sum of the digits of a factorial of numbers from 1 to 5 using list comprehension.
def factorial(n):
    if n == 0 or n==1:
        return 1
    else:
        return n * factorial(n-1)
sum_of_digits = [sum(int(digit) for digit in str(factorial(num))) for num in range(1, 6)]
print("Sum of digits of factorial of numbers from 1 to 5:",sum_of_digits)

#37. Write a program that finds the longest word in a sentence using list comprehension.
sentence = input("Enter a sentence: ")
longest_word = max((word.strip(".,?!") for word in sentence.split()), key=len)

print("Longest Word in the Sentence:", longest_word)

#38.Create a program that filters a list of strings to include only those with more than three vowels using list comprehension. 
def count_vowels(word):
    return sum(1 for char in word.lower() if char in 'aeiou')

strings_list = ["hello", "world", "programming", "python", "ai"]
filtered_strings = [word for word in strings_list if count_vowels(word) > 3]

print("Original List of Strings:", strings_list)
print("Filtered List (more than 3 vowels):", filtered_strings)

#39. Develop a program that calculates the sum of the digits of numbers from 1 to 1000 using list comprehension. 
sum_of_digits = sum(sum(int(digit) for digit in str(num)) for num in range(1, 1001))
print("Overall sum of digits from 1 to 1000:", sum_of_digits)

#40.Write a program that generates a list of prime palindromic numbers using list comprehension.
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def is_palindrome(num):
    return str(num) == str(num)[::-1]

prime_palindromic_numbers = [num for num in range(1, 1000) if is_prime(num) and is_palindrome(num)]
print("Prime Palindromic Numbers:", prime_palindromic_numbers)
```
