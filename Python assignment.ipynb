```python
#Explain the concept of recursion in Python. How does it differ from iteration?
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n*factorial(n-1)
num=10
result=factorial(num)
print(f"The factorial of {num} is {result}")

#Write a Python program to calculate the factorial of a number using recursion.
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        # Recursive case: n! = n*(n-1)!
        return n*factorial(n-1)
number=5
result=factorial(number)
print(f"The factorial of {number} is {result}")

#Create a recursive Python function to find the nth Fibonacci number.
def fibonacci(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        # Recursive case: Fibonacci(n) = Fibonacci(n-1) + Fibonacci(n-2)
        return fibonacci(n-1)+fibonacci(n-2)
n = 7
result = fibonacci(n)
print(f"The {n}th Fibonacci number is {result}")

#Write a recursive Python function to calculate the sum of all elements in a list.
def recursive_sum(lst):
    if not lst:
        return 0
    else:
        return lst[0] + recursive_sum(lst[1:])
my_list=[1,2,3,4,5]
result = recursive_sum(my_list)
print(f"The sum of the elements in the list is {result}")

#How can you prevent a recursive function from running indefinitely, causing a stack overflow error?
def fact(n):
    if n==0 or n==1:
        return 1
    else:
        # Recursive case: n! = n * (n-1)!
        return n * fact(n - 1)
result=fact(5)
print(result)

#Create a recursive Python function to find the greatest common divisor (GCD) of two numbers using the Euclidean algorithm.
def gcd(a,b):
    # Base case: GCD(a, 0) = a, GCD(0, b) = b
    if b==0:
        return a
    else:
        # Recursive case: GCD(a, b) = GCD(b, a % b)
        return gcd(b,a%b)
num1=48
num2=18
result=gcd(num1,num2)
print(f"The GCD of {num1} and {num2} is {result}")

#Write a recursive Python function to reverse a string.
def reverse_string(input_str):
    if len(input_str)<=1:
        return input_str
    else:
        return reverse_string(input_str[1:])+input_str[0]
original_str="Hello, World!"
reversed_str=reverse_string(original_str)
print(f"The original string: {original_str}")
print(f"The reversed string: {reversed_str}")

#Create a recursive Python function to calculate the power of a number (x^n).
def power(x,n):
    if n==0:
        return 1
    else:
        # Recursive case: x^n = x * x^(n-1)
        return x * power(x,n-1)
base=2
exponent=3
result=power(base,exponent)
print(f"{base} to the power of {exponent} is {result}")


#Write a recursive Python function to find all permutations of a given string.
def get_permutations(s):
    if len(s)==1:
        return [s]
    result = []
    for i in range(len(s)):
        first_char=s[i]
        remaining_chars=s[:i]+s[i+1:]
        permutations_of_remaining = get_permutations(remaining_chars)
        for perm in permutations_of_remaining:
            result.append(first_char+perm)  
    return result
input_string="abc"
permutations=get_permutations(input_string)
print(f"All permutations of '{input_string}': {permutations}")

#Write a recursive Python function to check if a string is a palindrome.
def is_palindrome(s):
    if len(s)<=1:
        return True
    else:
        return s[0]==s[-1] and is_palindrome(s[1:-1])
palindrome_str="radar"
non_palindrome_str="python"
result1=is_palindrome(palindrome_str)
result2=is_palindrome(non_palindrome_str)
print(f"Is '{palindrome_str}' a palindrome? {result1}")
print(f"Is '{non_palindrome_str}' a palindrome? {result2}")

#Create a recursive Python function to generate all possible combinations of a list of elements.
def generate_combinations(elements):
    if not elements:
        return [[]]
    first_element=elements[0]
    rest_of_elements=elements[1:]

    combinations_without_first=generate_combinations(rest_of_elements)

    combinations_with_first = [[first_element] + combo for combo in combinations_without_first]
    combinations_without_first.extend(combinations_with_first)

    return combinations_without_first
my_elements=[1,2,3]
all_combinations=generate_combinations(my_elements)
print(f"All combinations of {my_elements}: {all_combinations}")
```

```python
#What are lambda functions in Python, and when are they typically used?
"""Key characteristics of lambda functions:

Anonymous: Lambda functions are anonymous, meaning they don't have a name. They are usually used for short-term, specific purposes.

Single Expression: Lambda functions consist of a single expression, the result of which is implicitly returned.

Limited Functionality: Due to their simplicity, lambda functions are best suited for relatively simple operations. More complex tasks may be better handled by regular named functions."""
# Regular function
def square(x):
    return x**2
# Equivalent lambda function
square_lambda = lambda x:x**2

result1 = square(5)
result2 = square_lambda(5)
print(f"Using regular function: {result1}")
print(f"Using lambda function: {result2}")

#Write a Python program that uses lambda functions to sort a list of tuples based on the second element?
tuple_list=[(1,5),(3,2),(2,8), (4,1)]
sorted_list=sorted(tuple_list,key=lambda x:x[1])
print("Original list:",tuple_list)
print("Sorted list based on the second element:",sorted_list)

#Explain the concept of higher-order functions in Python, and provide an example.
def apply_operation(x,y,operation):
    """Applies the specified operation to the given arguments."""
    return operation(x,y)
def add(a,b):
    """Adds two numbers."""
    return a+b
def multiply(a,b):
    """Multiplies two numbers."""
    return a*b
result_addition=apply_operation(5,3,add)
result_multiplication=apply_operation(5,3,multiply)
print("Result of addition:", result_addition)
print("Result of multiplication:", result_multiplication)

def create_multiplier(factor):
    """Returns a function that multiplies its argument by the specified factor."""
    def multiplier(x):
        return x*factor
    return multiplier

double=create_multiplier(2)
triple=create_multiplier(3)
print("Double of 4:", double(4))
print("Triple of 4:", triple(4))

#Create a Python function that takes a list of numbers and a function as arguments, applying the function to each element in the list.
def apply_function_to_list(numbers,operation):
    """Applies the specified operation to each element in the list."""
    result=[]
    for num in numbers:
        result.append(operation(num))
    return result
numbers_list=[1,2,3,4,5]
def square(x):
    return x**2
squared_result=apply_function_to_list(numbers_list,square)
print("Original list:",numbers_list)
print("Squared list:",squared_result)
```
