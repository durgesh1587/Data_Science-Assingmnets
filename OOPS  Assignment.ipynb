```python
#Explain the concept of composition in Python and how it is used to build complex objects from simpler ones.
class Engine:
    def start(self):
        print("Engine started.")
    def stop(self):
        print("Engine stopped.")
class Car:
    def __init__(self):
        self.engine = Engine()
    def start(self):
        print("Starting the car...")
        self.engine.start()
    def stop(self):
        print("Stopping the car...")
        self.engine.stop()
my_car=Car()
my_car.start()  
my_car.stop() 

#Describe the difference between composition and inheritance in object-oriented programming.
"""Relationship Type:
Composition: In composition, classes are related by having instances of one class contained within another class. This forms a "has-a" relationship, where the containing class (composite) contains instances of other classes (components).
Inheritance: In inheritance, classes are related by forming an "is-a" relationship, where a subclass inherits attributes and methods from a superclass. The subclass is considered a specialization or subtype of the superclass.

Code Reuse:
Composition: Composition promotes code reuse by allowing objects of one class to be reused as components within another class. It encourages modular design and favors building objects by assembling simpler components.
Inheritance: Inheritance also promotes code reuse by allowing subclasses to inherit attributes and methods from a superclass. It facilitates the creation of hierarchies where subclasses share common functionality defined in the superclass.

Flexibility:
Composition: Composition offers more flexibility in object relationships and behavior. Since components are separate objects, they can be easily swapped or replaced without affecting the containing class. This promotes loose coupling and makes it easier to change the behavior of the composite object.
Inheritance: Inheritance establishes a tight coupling between subclasses and superclasses, as subclasses inherit the implementation details of the superclass. While inheritance supports code reuse and specialization, it can lead to tight coupling and hierarchical dependencies that make the codebase less flexible and more prone to changes.

Encapsulation:
Composition: Composition encourages encapsulation by allowing the containing class to control access to its component objects. It promotes information hiding and modular design by encapsulating the implementation details of components within the containing class.
Inheritance: Inheritance also supports encapsulation, but it can sometimes lead to tight coupling between classes, especially if subclasses rely heavily on the implementation details of the superclass. This can make it harder to maintain and refactor the codebase.

Complexity:
Composition: Composition tends to result in simpler and more modular code, as it allows complex objects to be built from simpler components. It promotes a "divide and conquer" approach to design, where complex systems are broken down into smaller, more manageable parts.
Inheritance: Inheritance can lead to more complex class hierarchies, especially in deep inheritance chains where subclasses inherit from multiple superclasses. It requires careful design to avoid issues such as the diamond problem and tight coupling between classes."""

#3. Create a Python class called `Author` with attributes for name and birthdate. Then, create a `Book` class that contains an instance of `Author` as a composition. Provide an example of creating a `Book` object.
class Author:
    def __init__(self,name,birthdate):
        self.name=name
        self.birthdate=birthdate
class Book:
    def __init__(self,title,author_name,author_birthdate):
        self.title=title
        self.author=Author(author_name, author_birthdate)
    def display_info(self):
        return f"Title: {self.title}\nAuthor: {self.author.name}\nBirthdate: {self.author.birthdate}"
author=Author("Durgesh","July 31, 1996")
book=Book("Harry Potter and the Philosopher's Stone", author.name, author.birthdate)
print(book.display_info())

#4. Discuss the benefits of using composition over inheritance in Python, especially in terms of code flexibility and reusability.
"""Flexibility:
Composition allows for more flexible and dynamic relationships between classes. With composition, you can easily change or replace the components of a class without modifying its behavior.
Unlike inheritance, which establishes a rigid hierarchy, composition enables a more loosely coupled design, where classes can be composed and recomposed in various configurations to meet changing requirements.

Code Reusability:
Composition promotes code reusability by encouraging the reuse of existing components in different contexts. Components can be designed to be modular and self-contained, making them easier to reuse in multiple classes or projects.
Since composition involves assembling objects from simpler components, it facilitates the reuse of those components across different classes, promoting a more modular and maintainable codebase.

Encapsulation:
Composition promotes encapsulation by allowing classes to control access to their component objects. Components are encapsulated within their containing classes, reducing the complexity of interactions between classes and promoting information hiding.
By encapsulating components, composition reduces the risk of unintended dependencies and side effects, leading to more robust and maintainable code.

Avoiding Tight Coupling:
Composition helps avoid tight coupling between classes, which can lead to brittle and inflexible designs. Since components are independent objects that interact through well-defined interfaces, changes to one component are less likely to affect others.
Inheritance, on the other hand, can create tight coupling between superclasses and subclasses, leading to hierarchical dependencies that are difficult to manage and refactor.

Promoting Design Patterns:
Composition facilitates the use of design patterns such as the Strategy pattern, Decorator pattern, and Dependency Injection pattern. These patterns leverage composition to achieve flexible and reusable designs by composing objects with different behaviors or responsibilities.
By embracing composition, developers can leverage these design patterns to address common design challenges and create more maintainable and extensible software systems."""

#How can you implement composition in Python classes? Provide examples of using composition to create complex objects.
class Engine:
    def start(self):
        print("Engine started.")
    def stop(self):
        print("Engine stopped.")
class Car:
    def __init__(self):
        self.engine = Engine()
    def start(self):
        print("Starting the car...")
        self.engine.start()
    def stop(self):
        print("Stopping the car...")
        self.engine.stop()
class Computer:
    def __init__(self):
        self.processor ="Intel i7" 
        self.memory ="16GB"     
        self.storage = "512GB SSD"
    def display_info(self):
        print("Computer specifications:")
        print(f"Processor: {self.processor}")
        print(f"Memory: {self.memory}")
        print(f"Storage: {self.storage}")
my_car=Car()
my_car.start()
my_car.stop()
print()
my_computer=Computer()
my_computer.display_info()

#6. Create a Python class hierarchy for a music player system, using composition to represent playlists and songs.
class Song:
    def __init__(self,title,artist,duration):
        self.title=title
        self.artist=artist
        self.duration=duration
    def play(self):
        print(f"Playing '{self.title}' by{self.artist} [{self.duration}]")
class Playlist:
    def __init__(self,name):
        self.name=name
        self.songs=[]
    def add_song(self,song):
        self.songs.append(song)
    def remove_song(self,song):
        if song in self.songs:
            self.songs.remove(song)
            print(f"Removed '{song.title}' from '{self.name}' playlist.")
        else:
            print(f"'{song.title}' is not in '{self.name}' playlist.")
    def play_all(self):
        print(f"Playing all songs in '{self.name}' playlist:")
        for song in self.songs:
            song.play()
        print()
song1=Song("Shape of You", "Ed Sheeran", "3:53")
song2=Song("Blinding Lights", "The Weeknd", "3:20")
song3=Song("Dance Monkey", "Tones and I", "3:30")

playlist1=Playlist("My Favorites")
playlist1.add_song(song1)
playlist1.add_song(song2)

playlist2=Playlist("Party Mix")
playlist2.add_song(song2)
playlist2.add_song(song3)

playlist1.play_all()
playlist2.play_all()

playlist1.remove_song(song1)
playlist1.play_all()

#Explain the concept of "has-a" relationships in composition and how it helps design software systems.
"""Component Class:
The class that is contained within another class is referred to as the component class or part class. It represents a distinct entity with its own attributes and behaviors.

Container Class: 
The class that contains instances of the component class is referred to as the container class or composite class. It encapsulates the component objects and provides methods to interact with them.

Composition: 
Composition is used to establish the "has-a" relationship between the container class and the component class. The container class contains one or more instances of the component class as attributes.

The "has-a" relationship helps design software systems by enabling a modular and flexible approach to building complex objects. Here are some key benefits:

Modularity: 
By using composition and "has-a" relationships, classes can be designed to represent individual components or parts of a system. This promotes modularity and code reuse by allowing components to be reused in different contexts.

Encapsulation: 
Composition promotes encapsulation by encapsulating the implementation details of components within the container class. Components are hidden from external classes, reducing dependencies and promoting information hiding.

Flexibility:
"Has-a" relationships allow for flexible and dynamic relationships between classes. Components can be easily added, removed, or replaced within the container class, allowing for changes in behavior or functionality without affecting the overall structure of the system.

Complexity Management: 
Composition helps manage the complexity of software systems by breaking them down into smaller, more manageable parts. Each part can be implemented and tested independently, leading to simpler and more maintainable code.

Hierarchical Structure:
Composition allows for the creation of hierarchical structures, where complex objects are built from simpler components. This facilitates the creation of complex systems by assembling smaller, reusable parts."""

#Create a Python class for a computer system, using composition to represent components like CPU, RAM, and storage devices.
class CPU:
    def __init__(self,brand,model,cores):
        self.brand=brand
        self.model=model
        self.cores=cores
    def info(self):
        return f"CPU: {self.brand} {self.model}, Cores: {self.cores}"
class RAM:
    def __init__(self,capacity_gb,speed_mhz):
        self.capacity_gb=capacity_gb
        self.speed_mhz=speed_mhz
    def info(self):
        return f"RAM: {self.capacity_gb}GB, Speed: {self.speed_mhz}MHz"
class Storage:
    def __init__(self,capacity_gb, type):
        self.capacity_gb=capacity_gb
        self.type=type
    def info(self):
        return f"Storage:{self.capacity_gb}GB {self.type}"
class Computer:
    def __init__(self,cpu,ram,storage):
        self.cpu=cpu
        self.ram=ram
        self.storage =storage
    def info(self):
        return f"Computer Specs:\n{self.cpu.info()}\n{self.ram.info()}\n{self.storage.info()}"
cpu=CPU("Intel", "Core i7",8)
ram=RAM(16,3200)
storage=Storage(512,"SSD")
computer=Computer(cpu,ram,storage)
print(computer.info())

#Describe the concept of "delegation" in composition and how it simplifies the design of complex systems.
"""Modularity and Reusability:
Delegation allows for breaking down complex systems into smaller, more manageable components. Each component is responsible for a specific set of tasks or behaviors, promoting modularity.
By delegating responsibilities to specialized objects, code becomes more reusable. These specialized objects can be reused in multiple contexts, reducing code duplication and promoting a more efficient development process.Separation of Concerns:

Delegation encourages the separation of concerns by assigning specific responsibilities to distinct objects. Each object focuses on a specific aspect of functionality, leading to cleaner and more maintainable code.
Separating concerns makes it easier to understand and reason about the behavior of individual components, as they are not burdened with unrelated responsibilities.

Encapsulation:
Delegation promotes encapsulation by allowing objects to encapsulate their behavior and delegate specific tasks to other objects. This hides the implementation details of delegated tasks, reducing complexity and preventing tight coupling between objects.
Encapsulation ensures that each object is responsible for managing its own state and behavior, leading to more robust and scalable code.

Flexibility and Extensibility:
Delegation enhances the flexibility and extensibility of complex systems by allowing components to be easily replaced or extended with minimal impact on other parts of the system.
Since each object delegates specific tasks to other objects, changes or enhancements can be made to individual components without affecting the overall system, making it easier to adapt to changing requirements.

Promotion of Design Patterns:
Delegation encourages the use of design patterns such as the Strategy pattern, Decorator pattern, and Adapter pattern. These patterns leverage delegation to achieve flexible and reusable designs by allowing objects to delegate responsibilities to other objects.
By embracing delegation and design patterns, developers can address common design challenges and create more maintainable and extensible software systems."""

#Create a Python class for a car, using composition to represent components like the engine, wheels, and transmission.
class Engine:
    def __init__(self,horsepower):
        self.horsepower=horsepower
    def start(self):
        print("Engine started.")
    def stop(self):
        print("Engine stopped.")
class Wheel:
    def __init__(self,size):
        self.size=size
    def rotate(self):
        print("Wheel rotating.")
class Transmission:
    def __init__(self,transmission_type):
        self.transmission_type=transmission_type
    def shift_gear(self,gear):
        print(f"Shifted to {gear} gear.")
class Car:
    def __init__(self,engine,wheels,transmission):
        self.engine=engine
        self.wheels=wheels
        self.transmission=transmission
    def start(self):
        self.engine.start()
    def stop(self):
        self.engine.stop()
    def drive(self):
        self.transmission.shift_gear("drive")
        print("Car is moving.")
    def park(self):
        self.transmission.shift_gear("park")
        print("Car is parked.")
engine=Engine(horsepower=200)
wheels=[Wheel(size=18) for _ in range(4)]
transmission=Transmission(transmission_type="automatic")
car=Car(engine,wheels,transmission)
car.start()
car.drive()
car.stop()
car.park()

#Create a Python class for a university course, using composition to represent students, instructors, and course materials.
class Student:
    def __init__(self,name,student_id):
        self.name=name
        self.student_id=student_id
    def __str__(self):
        return f"Student:{self.name}(ID:{self.student_id})"
class Instructor:
    def __init__(self,name,instructor_id):
        self.name=name
        self.instructor_id=instructor_id
    def __str__(self):
        return f"Instructor: {self.name} (ID:{self.instructor_id})"
class CourseMaterial:
    def __init__(self,textbook,slides):
        self.textbook=textbook
        self.slides =slides
    def __str__(self):
        return f"Course Material:\nTextbook: {self.textbook}\nSlides: {self.slides}"
class UniversityCourse:
    def __init__(self,course_name,instructor,students,course_material):
        self.course_name=course_name
        self.instructor=instructor
        self.students=students
        self.course_material=course_material
    def display_info(self):
        print(f"Course: {self.course_name}")
        print(f"Instructor: {self.instructor}")
        print("Students:")
        for student in self.students:
            print(student)
        print(f"Course Material:\n{self.course_material}")
student1=Student("monu",101)
student2=Student("Bola", 102)
instructor=Instructor("Dr. sonu", 201)
course_material=CourseMaterial("Introduction to Python", "Week 1 Slides")
students = [student1,student2]
course = UniversityCourse("Python Programming", instructor, students, course_material)
course.display_info()
    
#Create a Python class hierarchy for a restaurant system, using composition to represent menus, dishes,and ingredients.
class Ingredient:
    def __init__(self,name,quantity):
        self.name=name
        self.quantity=quantity
    def __str__(self):
        return f"{self.name}:{self.quantity}"
class Dish:
    def __init__(self,name,ingredients):
        self.name=name
        self.ingredients=ingredients
    def __str__(self):
        dish_info = f"Dish: {self.name}\nIngredients:\n"
        for ingredient in self.ingredients:
            dish_info+=f"-{ingredient}\n"
        return dish_info
class Menu:
    def __init__(self,name,dishes):
        self.name=name
        self.dishes=dishes
    def __str__(self):
        menu_info = f"Menu: {self.name}\n"
        for dish in self.dishes:
            menu_info+=str(dish) + "\n"
        return menu_info
ingredient1=Ingredient("Tomato","2 pieces")
ingredient2=Ingredient("Cheese", "100g")
ingredient3=Ingredient("Bread", "2 slices")
dish1=Dish("Grilled Cheese Sandwich", [ingredient1, ingredient2, ingredient3])
dish2=Dish("Margherita Pizza", [ingredient1, ingredient2])
menu=Menu("Lunch Menu", [dish1, dish2])
print(menu)

#15. Explain how composition enhances code maintainability and modularity in Python programs.?
"""Modularity:
Composition encourages breaking down complex systems into smaller, more manageable components or modules. Each component is responsible for a specific set of tasks or behaviors, promoting modularity.
By organizing code into modular components, it becomes easier to understand, debug, and maintain. Developers can focus on individual components independently, leading to a more organized and structured codebase.

Code Reusability:
Composition facilitates code reusability by promoting the creation of reusable components. These components can be used in multiple contexts or projects, reducing code duplication and promoting a more efficient development process.
By encapsulating functionality within components, developers can easily reuse them across different parts of the codebase, leading to a more maintainable and extensible system.

Encapsulation:
Composition promotes encapsulation by allowing objects to encapsulate their behavior and delegate specific tasks to other objects. This hides the implementation details of delegated tasks, reducing complexity and preventing tight coupling between objects.
Encapsulation ensures that each component is responsible for managing its own state and behavior, leading to more robust and scalable code.

Flexibility:
Composition enhances flexibility by allowing components to be easily replaced or extended with minimal impact on other parts of the system. Since each component is loosely coupled with others, changes to one component do not necessarily affect the entire system.
This flexibility makes it easier to adapt to changing requirements or add new features to the system without introducing unintended side effects or breaking existing functionality.

Testing and Debugging:
Composition simplifies testing and debugging by isolating components and making them easier to test in isolation. Each component can be tested independently, allowing for more targeted and effective testing strategies.
Debugging becomes more straightforward as issues are localized to specific components, making it easier to identify and fix problems without affecting other parts of the system."""

#Create a Python class for a computer game character, using composition to represent attributes like weapons, armor, and inventory.?
class Weapon:
    def __init__(self,name,damage):
        self.name=name
        self.damage=damage
    def __str__(self):
        return f"Weapon:{self.name},Damage:{self.damage}"
class Armor:
    def __init__(self,name,defense):
        self.name=name
        self.defense=defense
    def __str__(self):
        return f"Armor:{self.name},Defense:{self.defense}"
class Inventory:
    def __init__(self,items=None):
        self.items=items or []
    def add_item(self,item):
        self.items.append(item)
    def remove_item(self,item):
        if item in self.items:
            self.items.remove(item)
            print(f"Removed {item} from inventory.")
        else:
            print(f"{item} is not in the inventory.")
    def __str__(self):
        inventory_info = "Inventory:\n"
        for item in self.items:
            inventory_info+=f"-{item}\n"
        return inventory_info
class GameCharacter:
    def __init__(self,name,level,weapon=None,armor=None,inventory=None):
        self.name=name
        self.level=level
        self.weapon=weapon
        self.armor=armor
        self.inventory=inventory or Inventory()

    def __str__(self):
        return f"Character:{self.name}\nLevel:{self.level}\n{self.weapon}\n{self.armor}\n{self.inventory}"
sword=Weapon("Sword",10)
shield=Armor("Shield", 5)
inventory=Inventory()
inventory.add_item("Health Potion")
inventory.add_item("Mana Potion")
character = GameCharacter("Warrior",10,weapon=sword,armor=shield,inventory=inventory)
print(character)


#17. Describe the concept of "aggregation" in composition and how it differs from simple composition.
"""Independence of Lifetimes:
In aggregation, the part objects can exist independently of the whole object. This means that even if the whole object is destroyed, the part objects can still exist and be used elsewhere.
In simple composition, the part objects are owned and managed by the whole object. If the whole object is destroyed, the part objects are also destroyed.

Ownership:
In aggregation, the part objects are not owned by the whole object. They are typically passed to the whole object as references or pointers, but the whole object does not control their lifetime.
In simple composition, the part objects are owned and managed by the whole object. The whole object is responsible for creating, managing, and destroying the part objects.

Multiplicity:
Aggregation allows for a one-to-many relationship between the whole and its parts. The whole object can contain multiple instances of the part object(s).
Simple composition typically involves a one-to-one relationship between the whole and its parts. Each whole object has exactly one instance of the part object.

Flexibility:
Aggregation provides more flexibility because part objects can be shared among multiple whole objects. This allows for greater code reuse and modularity.
Simple composition tightly couples the whole object with its parts, making it less flexible and harder to reuse parts across different objects."""

#Create a Python class for a house, using composition to represent rooms, furniture, and appliances.
class Furniture:
    def __init__(self,name):
        self.name=name
    def __str__(self):
        return f"Furniture:{self.name}"
class Appliance:
    def __init__(self,name):
        self.name=name
    def __str__(self):
        return f"Appliance:{self.name}"
class Room:
    def __init__(self,name):
        self.name=name
        self.furniture=[]
        self.appliances=[]
    def add_furniture(self,furniture):
        self.furniture.append(furniture)
    def add_appliance(self,appliance):
        self.appliances.append(appliance)
    def __str__(self):
        room_info=f"Room:{self.name}\n"
        room_info+="Furniture:\n"
        for item in self.furniture:
            room_info+=f"-{item}\n"
        room_info +="Appliances:\n"
        for item in self.appliances:
            room_info+=f"-{item}\n"
        return room_info
class House:
    def __init__(self):
        self.rooms=[]
    def add_room(self,room):
        self.rooms.append(room)
    def __str__(self):
        house_info="House:\n"
        for room in self.rooms:
            house_info+=str(room)
        return house_info
living_room=Room("Living Room")
living_room.add_furniture(Furniture("Sofa"))
living_room.add_furniture(Furniture("Coffee Table"))
living_room.add_appliance(Appliance("Television"))
kitchen=Room("Kitchen")
kitchen.add_appliance(Appliance("Refrigerator"))
kitchen.add_appliance(Appliance("Microwave"))
house=House()
house.add_room(living_room)
house.add_room(kitchen)
print(house)

#19. How can you achieve flexibility in composed objects by allowing them to be replaced or modified dynamically at runtime?
"""Dependency Injection:
Dependency injection is a design pattern where dependencies of an object are provided externally rather than created internally. This allows for easy replacement of dependencies without modifying the object's code.
By injecting dependencies into objects, you can dynamically replace or modify the components of composed objects at runtime.

Interfaces and Abstract Classes:
Using interfaces or abstract classes to define the contract between components allows for interchangeable implementations.
You can dynamically switch implementations of interfaces or abstract classes at runtime, effectively replacing one component with another.

Factory Pattern:
The factory pattern encapsulates the creation of objects, allowing clients to obtain instances of objects without knowing the concrete implementation.
By using factories, you can dynamically select and create different implementations of components based on runtime conditions.

Dynamic Composition:
Dynamic composition involves constructing composed objects dynamically at runtime by adding or removing components as needed.
You can use techniques such as builder patterns or composite patterns to dynamically construct objects from smaller components, allowing for flexibility in composing objects.

Configuration Management:
Separating configuration from code allows for easy modification of object composition through configuration files or settings.
By externalizing configuration, you can change the composition of objects without modifying the code, providing flexibility in runtime behavior.

Aspect-Oriented Programming (AOP):
AOP allows for cross-cutting concerns such as logging, security, and error handling to be dynamically applied to objects at runtime.
By dynamically applying aspects to objects, you can modify their behavior without changing their core functionality, enhancing flexibility.

Reflection and Metadata:
Reflection allows for introspection and modification of objects at runtime, enabling dynamic composition.
You can use reflection to inspect and manipulate objects, including their components and dependencies, providing flexibility in object composition."""

#20. Create a Python class for a social media application, using composition to represent users, posts, and comments.
class Comment:
    def __init__(self,user,content):
        self.user=user
        self.content=content
    def __str__(self):
        return f"{self.user}:{self.content}"
class Post:
    def __init__(self,user,content,comments=None):
        self.user=user
        self.content=content
        self.comments=comments or []
    def add_comment(self, comment):
        self.comments.append(comment)
    def __str__(self):
        post_info=f"{self.user} posted:\n{self.content}\nComments:\n"
        for comment in self.comments:
            post_info += f"-{comment}\n"
        return post_info
class User:
    def __init__(self, username):
        self.username=username
    def create_post(self,content):
        return Post(self,content)
    def comment_on_post(self,post,content):
        comment = Comment(self,content)
        post.add_comment(comment)
    def __str__(self):
        return self.username
user1=User("Alice")
user2=User("Bob")
post1=user1.create_post("Hello, world!")
user2.comment_on_post(post1,"Nice post!")
user1.comment_on_post(post1,"Thanks, Bob!")
print(post1)

```
