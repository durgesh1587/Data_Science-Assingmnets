```python
#21. Create a program that calculates the income tax based on the user's income and tax brackets.
def calculate_income_tax(income):
    brackets = [10000, 50000, 100000, float('inf')]
    rates = [0.1, 0.2, 0.3, 0.4]

    # Calculate income tax
    tax = 0
    previous_bracket = 0

    for i in range(len(brackets)):
        if income <= brackets[i]:
            tax += (income - previous_bracket) * rates[i]
            break
        else:
            tax += (brackets[i] - previous_bracket) * rates[i]
            previous_bracket = brackets[i]
            
    return tax

# Input user income
try:
    user_income = float(input("Enter your annual income: "))
except ValueError:
    print("Invalid input. Please enter a valid number.")
else:
    income_tax = calculate_income_tax(user_income)
    print(f"Your income tax is: ${income_tax:.2f}")

#22. Write a program that simulates a rock-paper-scissors game against the computer.    
import random

def get_user_choice():
    while True:
        user_choice = input("Enter your choice (rock, paper, or scissors): ").lower()
        if user_choice in ["rock", "paper", "scissors"]:
            return user_choice
        else:
            print("Invalid choice. Please enter rock, paper, or scissors.")

def get_computer_choice():
    return random.choice(["rock", "paper", "scissors"])

def determine_winner(user_choice, computer_choice):
    if user_choice == computer_choice:
        return "It's a tie!"
    elif (
        (user_choice == "rock" and computer_choice == "scissors") or
        (user_choice == "paper" and computer_choice == "rock") or
        (user_choice == "scissors" and computer_choice == "paper")
    ):
        return "You win!"
    else:
        return "Computer wins!"
    
while True:
    user_choice = get_user_choice()
    computer_choice = get_computer_choice()

    print(f"\nYou chose: {user_choice}")
    print(f"Computer chose: {computer_choice}")

    result = determine_winner(user_choice, computer_choice)
    print(result)

    play_again = input("\nDo you want to play again? (yes/no): ").lower()
    if play_again != "yes":
        print("Thanks for playing. Goodbye!")
        break

        
#23. Develop a program that generates a random password based on user preferences (length, complexity).        
import random
import string

#  generate a random password
def generate_password(length, include_lowercase, include_uppercase, include_digits, include_special_chars):
    chars = ""
    
    if include_lowercase:
        chars += string.ascii_lowercase
    if include_uppercase:
        chars += string.ascii_uppercase
    if include_digits:
        chars += string.digits
    if include_special_chars:
        chars += string.punctuation

    if not chars:
        return "Error: No character set selected. Please include at least one character set."

    password = ''.join(random.choice(chars) for _ in range(length))
    return password

try:
    length = int(input("Enter the desired password length: "))
    include_lowercase = input("Include lowercase letters? (yes/no): ").lower() == "yes"
    include_uppercase = input("Include uppercase letters? (yes/no): ").lower() == "yes"
    include_digits = input("Include digits? (yes/no): ").lower() == "yes"
    include_special_chars = input("Include special characters? (yes/no): ").lower() == "yes"
except ValueError:
    print("Invalid input. Please enter a valid number for password length.")
else:
    password = generate_password(length, include_lowercase, include_uppercase, include_digits, include_special_chars)
    print(f"\nGenerated Password: {password}")
    
#24. Create a program that implements a simple text-based adventure game with branching scenarios.
def start_game():
    print("Welcome to the Text-Based Adventure Game!")
    print("You wake up in a mysterious room. What do you do?")
    print("1. Look around")
    print("2. Open the door")

    choice = input("Enter your choice (1 or 2): ")

    if choice == "1":
        look_around()
    elif choice == "2":
        open_door()
    else:
        print("Invalid choice. Please enter 1 or 2.")
        start_game()

def look_around():
    print("\nYou look around the room and find a key on the table.")
    print("What do you do?")
    print("1. Take the key")
    print("2. Ignore the key and open the door")

    choice = input("Enter your choice (1 or 2): ")

    if choice == "1":
        take_key()
    elif choice == "2":
        open_door_with_no_key()
    else:
        print("Invalid choice. Please enter 1 or 2.")
        look_around()

def open_door():
    print("\nYou try to open the door, but it's locked.")
    print("What do you do?")
    print("1. Look around")
    print("2. Try to break the door")

    choice = input("Enter your choice (1 or 2): ")

    if choice == "1":
        look_around()
    elif choice == "2":
        break_door()
    else:
        print("Invalid choice. Please enter 1 or 2.")
        open_door()

def open_door_with_no_key():
    print("\nYou try to open the door, but it's still locked.")
    print("What do you do?")
    print("1. Look around")
    print("2. Try to break the door")

    choice = input("Enter your choice (1 or 2): ")

    if choice == "1":
        look_around()
    elif choice == "2":
        break_door()
    else:
        print("Invalid choice. Please enter 1 or 2.")
        open_door_with_no_key()

def take_key():
    print("\nYou take the key.")
    print("What do you do next?")
    print("1. Open the door using the key")
    print("2. Keep exploring the room")

    choice = input("Enter your choice (1 or 2): ")

    if choice == "1":
        open_door_with_key()
    elif choice == "2":
        look_around()
    else:
        print("Invalid choice. Please enter 1 or 2.")
        take_key()

def open_door_with_key():
    print("\nYou use the key to unlock the door.")
    print("Congratulations! You have escaped from the mysterious room.")

def break_door():
    print("\nYou try to break the door, but it's too sturdy.")
    print("What do you do?")
    print("1. Look around")
    print("2. Keep trying to break the door")

    choice = input("Enter your choice (1 or 2): ")

    if choice == "1":
        look_around()
    elif choice == "2":
        break_door()
    else:
        print("Invalid choice. Please enter 1 or 2.")
        break_door()

# Start the game
start_game()


#25. Build a program that solves a linear equation for x, considering different cases.
def solve_linear_equation(a, b):
    #  equation ax + b = 0
    if a == 0:
        if b == 0:
            return "Infinite solutions (identity)"
        else:
            return "No solution (contradiction)"
    else:
        x = -b / a
        return f"Solution: x = {x}"

try:
    a = float(input("Enter the coefficient 'a' of the linear equation: "))
    b = float(input("Enter the constant term 'b' of the linear equation: "))
except ValueError:
    print("Invalid input. Please enter valid numerical coefficients.")
else:
    result = solve_linear_equation(a, b)
    print(result)

#26. Write a program that simulates a basic quiz game with multiple-choice questions and scoring.
def solve_linear_equation(a, b):
    # Solve for x in the equation ax + b = 0
    if a == 0:
        if b == 0:
            return "Infinite solutions (identity)"
        else:
            return "No solution (contradiction)"
    else:
        x = -b / a
        return f"Solution: x = {x}"

# Input coefficients a and b
try:
    a = float(input("Enter the coefficient 'a' of the linear equation: "))
    b = float(input("Enter the constant term 'b' of the linear equation: "))
except ValueError:
    print("Invalid input. Please enter valid numerical coefficients.")
else:
    result = solve_linear_equation(a, b)
    print(result)

#27.Develop a program that determines whether a given year is a prime number or not. 
def is_prime(year):
    if year <= 1:
        return False  

    for i in range(2, int(year**0.5) + 1):
        if year % i == 0:
            return False  

    return True 
try:
    year = int(input("Enter a year to check if it's a prime number: "))
except ValueError:
    print("Invalid input. Please enter a valid year.")
else:
    if is_prime(year):
        print(f"{year} is a prime number.")
    else:
        print(f"{year} is not a prime number.")
        
#28. Create a program that sorts three numbers in ascending order using conditional statements.        
try:
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    num3 = float(input("Enter the third number: "))
except ValueError:
    print("Invalid input. Please enter valid numerical values.")
else:
    if num1 <= num2 <= num3:
        sorted_nums = [num1, num2, num3]
    elif num1 <= num3 <= num2:
        sorted_nums = [num1, num3, num2]
    elif num2 <= num1 <= num3:
        sorted_nums = [num2, num1, num3]
    elif num2 <= num3 <= num1:
        sorted_nums = [num2, num3, num1]
    elif num3 <= num1 <= num2:
        sorted_nums = [num3, num1, num2]
    else:
        sorted_nums = [num3, num2, num1]
    print(f"The numbers in ascending order: {sorted_nums}")
    
#29. Build a program that determines the roots of a quartic equation using numerical methods.    
def quartic_equation(x, coefficients):
    # Evaluate the quartic equation for a given x
    return sum(coef * x ** (4 - i) for i, coef in enumerate(coefficients))

def quartic_derivative(x, coefficients):
    # Evaluate the derivative of the quartic equation for a given x
    return sum((4 - i) * coef * x ** (3 - i) for i, coef in enumerate(coefficients[:-1]))

def newton_method(initial_guess, coefficients, tolerance=1e-6, max_iterations=100):
    # Newton's method for finding roots
    x = initial_guess
    for _ in range(max_iterations):
        fx = quartic_equation(x, coefficients)
        dfx = quartic_derivative(x, coefficients)

        if abs(fx) < tolerance:
            return x  # Found a root within tolerance

        x = x - fx / dfx

    return None  # No convergence within the maximum number of iterations

# Coefficients of the quartic equation: ax^4 + bx^3 + cx^2 + dx + e
coefficients = [1, -10, 35, -50, 24]

# Find roots using Newton's method with initial guesses
roots = []
initial_guesses = [1.0, 2.0, 3.0, 4.0]
for guess in initial_guesses:
    root = newton_method(guess, coefficients)
    if root is not None:
        roots.append(root)

# Print the roots
if roots:
    print("Approximate roots:", roots)
else:
    print("No roots found within the specified tolerance.")

#30. Write a program that calculates the BMI (Body Mass Index) and provides health recommendations based
 def calculate_bmi(weight, height):
    # Formula to calculate BMI: BMI = weight (kg) / (height (m) * height (m))
    bmi = weight / (height ** 2)
    return bmi

def get_health_recommendation(bmi):
    if bmi < 18.5:
        return "Underweight: Consider gaining some weight. Consult with a healthcare professional."
    elif 18.5 <= bmi < 24.9:
        return "Normal weight: Maintain a healthy lifestyle with balanced nutrition and regular exercise."
    elif 25 <= bmi < 29.9:
        return "Overweight: Consider losing some weight. Consult with a healthcare professional."
    else:
        return "Obese: It is recommended to focus on weight loss. Consult with a healthcare professional."

# Input user weight and height
try:
    weight = float(input("Enter your weight in kilograms: "))
    height = float(input("Enter your height in meters: "))
except ValueError:
    print("Invalid input. Please enter valid numeric values.")
else:
    # Calculate BMI
    bmi = calculate_bmi(weight, height)
    
    # Get health recommendation based on BMI
    recommendation = get_health_recommendation(bmi)

    # Display BMI and health recommendation
    print(f"Your BMI is: {bmi:.2f}")
    print(f"Health Recommendation: {recommendation}")
  
```
